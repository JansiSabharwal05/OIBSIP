```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
file1 = '/content/Unemployment_in_India[1].csv'
file2 = '/content/Unemployment_Rate_upto_11_2020[1].csv'

df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
```

```python
print("Dataset 1: Unemployment in India")
print(df1.head())
print("\nDataset 2: Unemployment Rate up to 11/2020")
print(df2.head())
```

```python
# Remove leading/trailing spaces from column names
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()

# Now proceed with the original code:
df1['Date'] = pd.to_datetime(df1['Date'])
df2['Date'] = pd.to_datetime(df2['Date'])
```

```python
df1.set_index('Date', inplace=True)
df2.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=df2, x=df2.index, y='Estimated Unemployment Rate (%)', label='Unemployment Rate')
plt.axvline(pd.to_datetime('2020-03-01'), color='r', linestyle='--', label='COVID-19 Start')
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.title("Unemployment Rate in India Over Time")
plt.legend()
plt.grid(True)
plt.show()
```
